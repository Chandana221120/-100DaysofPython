```python
# Algorith approach
# Start -> Generate random word -> Generate as many blanks as in word -> Ask user to guess a letter -> check if letter in word? 
# If yes -> Replace the blank with the letter -> Are all blanks filled? -> No -> back to ask user -> yes -> Game over
# If no -> Lose a life -> Have they run out of lives? -> No -> back to ask user -> yes -> Game over

```

```python
# Step1: Choosing random word from a list
words = ["ardvark","baboon","camel"]
import random
a = random.randint(0, len(words)-1)
chosen_word = words[a]
print(chosen_word)

# Choosing random word - Method_2
chosen_word2 = random.choice(words)
print(chosen_word2)

# Ste2: Ask user to guess a letter, assign to variable called guess. Make guess lowercase
guess = input("Guess a letter : ").lower()

# Step3: Check if letter the user guessed is one of letters in chosen_word
for i in chosen_word:  
    if i == guess:
        print("right")
    else:
        print("wrong")
```

```python
# Step4: Create empty list called Display.For each letter in chosen word, add "_".
display = []
value = "_"
for I in range(len(chosen_word)):
    display += "_"
display
```

```python
# Step5: loop through each position in chosen_word, if letter matches position, reveal
for i in range(len(chosen_word)):
    if chosen_word[i] == guess:
        display[i] = guess
print(display)
```

```python
# Step6: Use while loop to let the user guess again, Loop should stop after all letters are guessed.

display_check = ''.join(display)
print(display_check)

end_of_game = False

while not end_of_game:
    guess = input("Guess a letter : ").lower()
    for i in range(len(chosen_word)):
        if chosen_word[i] == guess:
            display[i] = guess
    print("Chose word is: ", chosen_word)                
    print("Display is: " , ''.join(display))
    end_of_game = chosen_word == ''.join(display)
else:
    print("You won!")
```

```python
# Step7: Create varible called lives to keep track of number of lives left. Set 'lives' to 6. If it goes to 0, stop game.
words = ["ardvark","baboon","camel"]
import random

chosen_word = random.choice(words)
print(chosen_word)

display = []
value = "_"
for I in range(len(chosen_word)):
    display += "_"
display

end_of_game = False
lives = 6

while not end_of_game:
    guess = input("Guess a letter : ").lower()
    
    if guess in display:
        print(f"You've already guessed {guess}")
        
    for i in range(len(chosen_word)):
        if chosen_word[i] == guess:
            display[i] = guess
    if guess not in chosen_word:
        lives -= 1
        if lives == 0:
            end_of_game = True
            print("No lives left!")
    if "_" not in display:
        end_of_game = True
        print("You win!")
        
    print("Chose word is: ", chosen_word)                
    print("Display is: " , ''.join(display))
    print("Lives left: ", lives)

```
